# Importar Librerías

```python
import numpy as np
import os
import re
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
import keras
import tensorflow as tf
from tensorflow.keras import Sequential, layers
from keras.utils import to_categorical
from keras.models import Sequential, Model
from keras.layers import Input
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D
from keras.layers import BatchNormalization
from keras.layers import LeakyReLU
from keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
```

# Cargar set de Imágenes

```python
imgpath = r"C:\Users\rival\Downloads\Dataset\Dataset_85x85"
#imgpath = "/mnt/c/Users/rival/Downloads/Dataset/Dataset_85x85"

images = []
directories = []
dircount = []
prevRoot=''
cant=0

print("Leyendo imagenes de:", imgpath)

# Escaneo de la estructura de carpetas
for root, dirnames, filenames in os.walk(imgpath):
    # Filtrar solo archivos con extensiones de imagen validas
    # Usamos r"\." para evitar el SyntaxWarning del escape sequence
    valid_images = [f for f in filenames if re.search(r"\.(jpg|jpeg|png|bmp|tiff)$", f, re.IGNORECASE)]
    
    # Si la carpeta no tiene imagenes, saltar a la siguiente
    if not valid_images:
        continue

    print(f"Entrando a la carpeta: {root}")
    directories.append(root)
    cant_actual = 0

    for filename in valid_images:
        filepath = os.path.join(root, filename)
        try:
            # Leer imagen
            image = plt.imread(filepath)
            images.append(image)
            cant_actual += 1
            print(f"Imagenes cargadas: {cant_actual}", end="\r")
        except Exception as e:
            print(f"Error al leer el archivo {filename}: {e}")

    # Guardar el total de la carpeta despues de terminar el bucle interno
    dircount.append(cant_actual)
    print(f"\nFinalizado: {os.path.basename(root)} tiene {cant_actual} imagenes.")

print("\n--- Resumen del Conjunto de Datos ---")
print('Total de directorios leidos:', len(directories))
print("Distribucion por categoria:", list(zip([os.path.basename(d) for d in directories], dircount)))
print('Suma total de imagenes cargadas en memoria:', sum(dircount))
```

# Creamos las etiquetas

```python
labels=[]
indice=0
for cantidad in dircount:
    for i in range(cantidad):
        labels.append(indice)
    indice=indice+1
print("Cantidad etiquetas creadas: ",len(labels))
```

```python
deportes=[]
indice=0
for directorio in directories:
    name = directorio.split(os.sep)
    print(indice , name[len(name)-1])
    deportes.append(name[len(name)-1])
    indice=indice+1
```

```python
y = np.array(labels)
X = np.array(images, dtype=np.uint8) #convierto de lista a numpy

# Find the unique numbers from the train labels
classes = np.unique(y)
nClasses = len(classes)
print('Total number of outputs : ', nClasses)
print('Output classes : ', classes)
```

# Creamos Sets de Entrenamiento y Test

```python
train_X,test_X,train_Y,test_Y = train_test_split(X,y,test_size=0.2)
print('Training data shape : ', train_X.shape, train_Y.shape)
print('Testing data shape : ', test_X.shape, test_Y.shape)
```

```python
plt.figure(figsize=[5,5])

# Display the first image in training data
plt.subplot(121)
plt.imshow(train_X[0,:,:], cmap='gray')
plt.title("Ground Truth : {}".format(train_Y[0]))

# Display the first image in testing data
plt.subplot(122)
plt.imshow(test_X[0,:,:], cmap='gray')
plt.title("Ground Truth : {}".format(test_Y[0]))
```

# Preprocesamos las imagenes

```python
train_X = train_X.astype('float32')
test_X = test_X.astype('float32')
train_X = train_X / 255.
test_X = test_X / 255.
```

# Creamos el Set de Entrenamiento y Validación

```python
#Mezclar todo y crear los grupos de entrenamiento y testing
train_X, valid_X, train_label, valid_label = train_test_split(
    train_X, train_Y, test_size=0.2, random_state=13
)

train_label = train_label.astype(np.int32)
valid_label = valid_label.astype(np.int32)
```

```python
print(train_X.shape,valid_X.shape,train_label.shape,valid_label.shape)
```

# Creamos el modelo de CNN

```python
#declaramos variables con los parámetros de configuración de la red
INIT_LR = 1e-2 # Valor inicial de learning rate. El valor 1e-3 corresponde con 0.001
epochs = 100 # Cantidad de iteraciones completas al conjunto de imagenes de entrenamiento
batch_size = 16 # cantidad de imágenes que se toman a la vez en memoria
```

```python
sport_model = Sequential()
# Bloque 1
sport_model.add(Conv2D(32, kernel_size=(3, 3), activation='linear', padding='same', input_shape=(85,85,3)))
sport_model.add(LeakyReLU(alpha=0.1))
sport_model.add(BatchNormalization())
sport_model.add(MaxPooling2D(pool_size=(2, 2)))
sport_model.add(Dropout(0.25))

# Bloque 2
sport_model.add(Conv2D(64, kernel_size=(3, 3), activation='linear', padding='same'))
sport_model.add(LeakyReLU(alpha=0.1))
sport_model.add(BatchNormalization())
sport_model.add(MaxPooling2D(pool_size=(2, 2)))
sport_model.add(Dropout(0.25))

# Bloque 3
sport_model.add(Conv2D(128, kernel_size=(3, 3), activation='linear', padding='same'))
sport_model.add(LeakyReLU(alpha=0.1))
sport_model.add(BatchNormalization())
sport_model.add(MaxPooling2D(pool_size=(2, 2)))
sport_model.add(Dropout(0.4))

# Clasificación final
sport_model.add(Flatten())
sport_model.add(Dense(64, activation='linear'))
sport_model.add(LeakyReLU(alpha=0.1))
sport_model.add(Dropout(0.5))
sport_model.add(Dense(nClasses, activation='softmax'))
```

```python
sport_model.summary()
```

```python
sport_model.compile(
    loss='sparse_categorical_crossentropy',
    optimizer=keras.optimizers.Adagrad(learning_rate=INIT_LR),
    metrics=['accuracy']
)
```

# Entrenamos el modelo: Aprende a clasificar imágenes

```python
# este paso puede tomar varios minutos, dependiendo de tu ordenador, cpu y memoria ram libre
# como ejemplo, en mi Macbook pro tarda 4 minutos
early_stop=EarlyStopping(monitor='val_loss', patience=20, restore_best_weights=True, verbose=1)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, min_lr=1e-7, verbose=1)
checkpoint= ModelCheckpoint('best_model.keras', monitor='val_loss', save_best_only=True, verbose=1)
sport_train = sport_model.fit(train_X, train_label, batch_size=batch_size,epochs=epochs,verbose=1,validation_data=(valid_X, valid_label), 
                             callbacks=[early_stop, reduce_lr, checkpoint])
```

```python
# guardamos la red, para reutilizarla en el futuro, sin tener que volver a entrenar
sport_model.save("animales.keras")
```

# Evaluamos la red

```python
test_Y = test_Y.astype(np.int32)
test_eval = sport_model.evaluate(test_X, test_Y, verbose=1)
```

```python
print('Test loss:', test_eval[0])
print('Test accuracy:', test_eval[1])
```

```python
accuracy = sport_train.history['accuracy']
val_accuracy = sport_train.history['val_accuracy']
loss = sport_train.history['loss']
val_loss = sport_train.history['val_loss']
epochs = range(len(accuracy))
plt.plot(epochs, accuracy, 'bo', label='Training accuracy')
plt.plot(epochs, val_accuracy, 'b', label='Validation accuracy')
plt.title('Training and validation accuracy')
plt.legend()
plt.figure()
plt.plot(epochs, loss, 'bo', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.legend()
plt.show()
```

# Prediccion de una nueva imagen

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.transform import resize
from tensorflow.keras.models import load_model

sport_model = load_model("animales.keras")

deportes = ['gatos', 'hormigas', 'mariquitas', 'perros', 'tortugas']  

# Lista de imágenes a probar
filenames = [r'C:\Users\rival\Downloads\Dataset\Test\hormiga.jpg']

images = []

for filepath in filenames:
    image = plt.imread(filepath)
    
    image_resized = resize(
        image, (85, 85), 
        anti_aliasing=True,
        clip=False,
        preserve_range=True
    )
    
    if len(image_resized.shape) == 2:
        image_resized = np.stack((image_resized,)*3, axis=-1)
    
    images.append(image_resized)

X = np.array(images, dtype=np.uint8)

test_X = X.astype('float32') / 255.0

print("Forma de test_X:", test_X.shape)

predicted_classes = sport_model.predict(test_X)

for i, img_tagged in enumerate(predicted_classes):
    predicted_index = np.argmax(img_tagged)
    print(f"{filenames[i]} -> {deportes[predicted_index]}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.transform import resize
from tensorflow.keras.models import load_model
import os
from pathlib import Path

# Cargar el modelo
sport_model = load_model("animales.keras")
deportes = ['gato', 'hormigas', 'mariquitas', 'perros', 'tortugas']

carpeta_imagenes = r"C:\Users\rival\Downloads\Dataset\Test"

# Obtener todas las imágenes de la carpeta
extensiones_validas = ('.jpg', '.jpeg', '.png', '.bmp', '.tiff', '.tif')
filenames = []

for archivo in os.listdir(carpeta_imagenes):
    if archivo.lower().endswith(extensiones_validas):
        filenames.append(os.path.join(carpeta_imagenes, archivo))

# Verificar si se encontraron imágenes
if not filenames:
    print(f"No se encontraron imágenes en la carpeta '{carpeta_imagenes}'")
else:
    print(f"Se encontraron {len(filenames)} imágenes para procesar\n")
    
    images = []
    for filepath in filenames:
        try:
            image = plt.imread(filepath)
            
            image_resized = resize(
                image, (85, 85), 
                anti_aliasing=True,
                clip=False,
                preserve_range=True
            )
            
            # Convertir imagen en escala de grises a RGB si es necesario
            if len(image_resized.shape) == 2:
                image_resized = np.stack((image_resized,)*3, axis=-1)
            
            images.append(image_resized)
        except Exception as e:
            print(f"Error al procesar {filepath}: {e}")
            filenames.remove(filepath)
    
    # Procesar solo si hay imágenes válidas
    if images:
        X = np.array(images, dtype=np.uint8)
        test_X = X.astype('float32') / 255.0
        print("Forma de test_X:", test_X.shape)
        print("\nResultados de la clasificación:")
        print("-" * 50)
        
        predicted_classes = sport_model.predict(test_X)
        
        for i, img_tagged in enumerate(predicted_classes):
            predicted_index = np.argmax(img_tagged)
            confianza = img_tagged[predicted_index] * 100
            nombre_archivo = os.path.basename(filenames[i])
            print(f"{nombre_archivo:30} -> {deportes[predicted_index]:10} ({confianza:.2f}%)")
    else:
        print("No se pudieron procesar imágenes válidas")
```
